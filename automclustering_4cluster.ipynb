```python
import numpy as np 
import matplotlib.pyplot as plt 

# random number generation 
# 2-D data 
num_samples=1000

X = np.random.randint(10,100,size=(num_samples))
Y = np.random.randint(10,100,size=(num_samples))

# let's visualise the 2D data 
plt.figure(1)
plt.plot(X,Y,'ro')
# markers - o-circle, d-diamond, *-star,s - square

plt.title('Random Raw Data')
print(X.shape)
print(Y.shape)
```

```python
# first step 
# take the datapoints as centroid

C1x = X[10]; C1y=Y[10] # coor. of centroid 1
C2x = X[11]; C2y=Y[11] # coor. of centroid 2
C3x = X[12]; C3y=Y[12] # coor. of centroid 1
C4x = X[13]; C4y=Y[13] # coor. of centroid 2


for it in range(30):
    # second step
    cluster1x=[];cluster1y=[]
    cluster2x=[];cluster2y=[]
    cluster3x=[];cluster3y=[]
    cluster4x=[];cluster4y=[]

    for i in range(num_samples):
        d1 =  np.sqrt((C1x-X[i])**2 + (C1y-Y[i])**2)
        d2 =  np.sqrt((C2x-X[i])**2 + (C2y-Y[i])**2)
        d3 =  np.sqrt((C3x-X[i])**2 + (C3y-Y[i])**2)
        d4 =  np.sqrt((C4x-X[i])**2 + (C4y-Y[i])**2)

        if(d1<d2 and d1<d3 and d1<d4):
            cluster1x.append(X[i])
            cluster1y.append(Y[i])
            plt.plot(X[i],Y[i],'bd')
        elif(d2<d1 and d2<d3 and d2<d4):
            cluster2x.append(X[i])
            cluster2y.append(Y[i])
            plt.plot(X[i],Y[i],'rs')
        elif(d3<d1 and d3<d2 and d3<d4):
            cluster3x.append(X[i])
            cluster3y.append(Y[i])
            plt.plot(X[i],Y[i],'ms')
        elif(d4<d1 and d4<d3 and d4<d2):
            cluster4x.append(X[i])
            cluster4y.append(Y[i])
            plt.plot(X[i],Y[i],'cs')
            
    plt.plot(C1x,C1y,'ks')
    plt.plot(C2x,C2y,'ks')
    plt.plot(C3x,C3y,'ks')
    plt.plot(C4x,C4y,'ks')
    C1xavg = sum(cluster1x)/len(cluster1x)
    C1yavg = sum(cluster1y)/len(cluster1y)
    C2xavg = sum(cluster2x)/len(cluster2x)
    C2yavg = sum(cluster2y)/len(cluster2y)
    C3xavg = sum(cluster3x)/len(cluster3x)
    C3yavg = sum(cluster3y)/len(cluster3y)
    C4xavg = sum(cluster4x)/len(cluster4x)
    C4yavg = sum(cluster4y)/len(cluster4y)
    C1x = C1xavg;C1y = C1yavg
    C2x = C2xavg;C2y = C2yavg
    C3x = C3xavg;C3y = C3yavg
    C4x = C4xavg;C4y = C4yavg
    plt.title(['Iteration: ',it+1])
    
    plt.pause(0.3) # 0.3 seconds delay after each complete iteration 


        
        
        
```
